## Abstract
<a id="abstract"></a>

Lung Cancer remains a leading cancer-related deaths globaly, mainly due to the lack of an accurate and timely diagnosis. This project tackles ths problem by developing a robust Computer-Aided Diagnosis (CAD) system for pulmonary tumour malignancy classification, using Computed Tomography (CT) data from the LIDC-IDRI dataset.

The proposed solution employes a hybrid feature extraction methodology, incorporating deep feature extraction using the pre-trained neural network **DenseNet** and the **PyRadiomics** library, with the hope to deliver both satisfactory performance and good clinical interpretability. This feature set is then optimized for dimensionality using Principal Component Analysis (PCA). Then, the Bag of Features (BoF) technique is applied in order to create compact and reduce dimensionality while ensuring the representation is order-independent. This new data set is kept and used separately for benchmarking purposes.

Finally, two of the same classical machine learning model, are trained on one of the data sets (with or without BoF) and they are compared against eachother, to determine which is more performant and, therefore, should be used in the final system. Here we implement various models, namely AdaBoost, Logistic Regression, Random Forest, SVM, and XGBoost. They are then benchmarked using their Recall and AUC scores (as well as other metrics deemed relevant), being the most appropriate chosen to integrate the final system.

The resulting system aims to serve as a reliable second opinion for radiologists, reducing inter-observer variability and accelerating the triage process. This technological innovation is expected to significantly improve public health outcomes by supporting the early detection of malignancy, while adhering to strict regulatory requirements for medical devices and high-risk AI.

## Introduction

Visual assessment of pulmonary nodules in Computed Tomography (CT) scans by radiologists is hindered by high inter-observer variability and the large volume of data. This clinical limitation drives the development of Artificial Intelligence (AI)-based Computer-Aided Diagnosis (CAD) Systems to provide an objective and automatic assessment of malignancy risk.

Previous work has demonstrated the potential of Machine Learning and Deep Learning solutions for classifying pulmonary nodules using databases such as LIDC-IDRI [[AMB+15](https://www.cancerimagingarchive.net/collection/lidc-idri/)]. The resulting feature set will be optimized using Principal Component Analysis (PCA) and classified by a Random Forest model. This approach aims to provide a CAD system that acts as a reliable second opinion, optimizing the Sensitivity (Recall) and AUC metric to ensure the minimization of false negatives in screening, while maintaining the transparency required for clinical and regulatory adoption.

### Feature extraction
<a id="feature-extraction"></a>
One of the most important steps when using classical ML methods is *feature extraction*: transforming raw image or volumetric data into informative, hand-crafted or engineered features that can effectively discriminate malignant from benign nodules. In this chapter, we motivate why feature extraction is a key component of the project’s benchmark experiments and review prior works that have successfully used such methods on LIDC-IDRI.

#### Why Feature Extraction Is Important in This Project

1. **Data attributes in LIDC-IDRI**  
   The **[LIDC-IDRI dataset](https://wiki.cancerimagingarchive.net/display/Public/LIDC-IDRI)** provides a unique foundation for feature extraction because of its comprehensive radiologist annotations.  
   Each nodule is independently evaluated by up to four radiologists, who provide both **segmentation masks** and **subjective ratings** of visual features including *shape, margin, spiculation, lobulation, texture, calcification, and sphericity*.  
   These annotations allow for the extraction of both **semantic** (annotation-based) and **quantitative** (voxel-based) features, supporting a dual-layer analysis approach consistent with prior studies (Hancock & Magnan, *Medical Physics*, 2017; Shen et al., *IEEE Transactions on Medical Imaging*, 2015).

2. **3D and 2D feature extraction capabilities**  
   A distinctive advantage of LIDC-IDRI is that it enables feature extraction in both **2D** and **3D** domains:
   - In **2D**, features can be computed slice-by-slice, capturing local variations in intensity or texture particularly useful when comparing to conventional radiological interpretation, which often relies on axial slices.  
   - In **3D**, features capture the complete volumetric representation of the nodule, allowing accurate measurement of global shape, volume, and spatial texture heterogeneity.  
   
   As reported by Fave et al. (*Medical Physics*, 2015) and Aerts et al. (*Nature Communications*, 2014), 3D radiomic descriptors derived from isotropically resampled CT scans provide significantly more robust and reproducible biomarkers than their 2D counterparts.
 

#### Use of the PyLIDC Library for Feature Extraction
<a id="dicom"></a>
A key component that enabled efficient and structured access to the LIDC-IDRI dataset in this project was the **[PyLIDC](https://pylidc.github.io/)** library. PyLIDC is a Python package specifically designed to interface with the LIDC-IDRI lung CT dataset, simplifying the management of volumetric scans, annotations, and segmentation masks. It abstracts much of the manual file handling, parsing, and DICOM metadata navigation that would otherwise be necessary.

Instead of reading and organizing thousands of DICOM files manually, PyLIDC stores metadata about patients, scans, and annotations in a local **SQL-based database**. This architecture allows for highly efficient querying of nodules and their related attributes, such as radiologist annotations, malignancy scores, and segmentation masks. Queries that would normally require multiple filesystem traversals can thus be performed instantly with simple Python commands.


##### Core Classes Used: `Scan` and `Annotation`

Two main PyLIDC classes were extensively used during the feature extraction stage:

1. **`Scan` class**  
   The `Scan` object represents a single CT scan of a patient.  
   It provides methods to:
   - Retrieve the list of DICOM slices that compose the 3D volume (`get_path_to_dicom_files()`).
   - Reconstruct the full 3D image volume as a NumPy array (`to_volume()`).
   - Access the voxel spacing and orientation metadata, crucial for spatial normalization.
   - Query the set of associated annotations made by radiologists (`scan.annotations`).

   This abstraction is essential for loading scans consistently, handling voxel anisotropy, and preparing volumes for radiomics feature computation.

2. **`Annotation` class**  
   Each `Annotation` represents a single radiologist’s interpretation of a nodule within a scan.  
   It includes:
   - The **binary segmentation mask** of the nodule region (`annotation.boolean_mask()`).
   - The **3D bounding box coordinates** for the segmented volume (`annotation.bbox()`).
   - Radiologist-assigned attributes such as **malignancy**, **spiculation**, **texture**, **margin**, **sphericity**, and other qualitative ratings.
   - Aggregation utilities to combine multiple annotations into a consensus representation.

   These attributes were fundamental to extracting both **quantitative (shape, intensity, texture)** and **semantic (radiologist ratings)** features for machine learning classification.

#### Use of PyRadiomics for 3D Feature Extraction
<a id="pyradiomics"></a>
To complement the semantic features provided by the radiologists’ annotations, the project also employed the **[PyRadiomics](https://pyradiomics.readthedocs.io/en/latest/)** library to extract a large set of quantitative, high-dimensional features directly from the 3D CT volumes of each nodule.  
PyRadiomics is a well-established open-source toolkit for **radiomic feature extraction**, supporting reproducible measurement of intensity, shape, and texture statistics from medical imaging data.

This approach allowed us to perform a **3D radiomic analysis** of each lung nodule enabled by the segmentation masks and spatial metadata obtained via the `Scan` and `Annotation` classes of PyLIDC. Without these tools, handling the spatial alignment, voxel spacing, and segmentation regions in LIDC-IDRI would have been prohibitively complex.


##### Pre-Radiomics workflow and use of NIFTI files

PyRadiomics operates primarily on medical image data in the **NIfTI (Neuroimaging Informatics Technology Initiative)** format a widely used file format (`.nii` or `.nii.gz`) that stores volumetric data along with spatial information (voxel size, orientation, and affine transformations).  
Using this format ensures interoperability with neuroimaging and radiology toolkits, allowing 3D CT data to be handled as volumetric tensors with correct real-world geometry.

To generate these NIfTI files, each annotated nodule from LIDC-IDRI was:

1. **Reconstructed** from the corresponding DICOM slices using PyLIDC’s `Scan.to_volume()` function.  
2. **Masked** by a **binary segmentation** (1 for nodule voxels, 0 for background).  
3. **Converted** to NIfTI format, producing two aligned files:  
   - The **image volume** (Hounsfield Unit values), and  
   - The **binary mask** (region of interest for radiomics computation).

This masking step was essential to avoid analyzing the entire CT volume, which contains millions of irrelevant voxels (≈4 million). By isolating the nodule region, the computation was reduced to about 40 000 voxels per nodule. This was a necessary optimization to prevent memory overflow and to ensure that extracted features describe the lesion only, not surrounding lung tissue.


##### Consensus level

To produce a single, reliable segmentation mask for radiomic extraction, we followed the **50% consensus threshold**, a commonly used strategy in state-of-the-art works (e.g., Armato et al., *Med. Phys.*, 2011; Zwanenburg et al., *Radiology*, 2020).  
This means that a voxel was included in the final mask if it was marked as part of the nodule by **at least half of the annotating radiologists**.

##### Voxel resampling

In addition to its role in ensuring comparability of radiomic features, the **resampling to isotropic voxel spacing of 1 × 1 × 1 mm** is also essential for the correct functioning of the **PyLIDC** data pipeline itself.

The LIDC-IDRI dataset was acquired across multiple institutions and scanners, leading to **heterogeneous voxel dimensions**, axial resolutions typically around [0.6 , 0.8]mm, but slice thicknesses ranging from [1.0 , 3.0]mm or more.  
When reconstructing nodules in 3D using PyLIDC’s `Scan.to_volume()` and `Annotation.boolean_mask()` methods, this anisotropy can distort the geometry of nodules along the z-axis. 

<br>

By resampling every scan and its associated annotation mask to **1 mm isotropic spacing**, we ensure that:

- **Geometric consistency is preserved:** 3D measurements such as volume, surface area, sphericity, and compactness represent the true physical shape of the nodule rather than scanner-dependent scaling.  
- **Interoperability with PyRadiomics is guaranteed:** PyRadiomics expects spatially uniform voxel spacing to compute gray-level matrices (GLCM, GLSZM, etc.) consistently in all directions. Without isotropy, texture features could be directionally biased. 

As noted in prior radiomics studies (Aerts et al., *Nature Communications*, 2014; Fave et al., *Medical Physics*, 2015), isotropic resampling not only stabilizes texture-based features but also ensures that **database-driven frameworks like PyLIDC** can operate efficiently and consistently across diverse imaging sources.

#### Radiomics Workflow

The [nifti_atr.py](nifti_atr.py) file implements the feature extraction pipeline used in this project.  
Its design prioritizes the **extraction of the maximum number of quantitative features possible** from each nodule, leveraging the 3D masks and volumetric data provided by PyLIDC.

The rationale behind this exhaustive feature extraction is twofold:

1. **Enhanced exploratory data analysis (EDA):**  
   By generating a rich set of features: spanning intensity, shape, texture, and multiple filtered versions of the image, we ensure that the dataset contains sufficient descriptive information for identifying the most valuable patterns, correlations, and outliers among nodules. 

2. **Robust benchmarking:**  
  Comparing the diferent filters instead of picking just one ensures that the results represent most of the possible outcomes for this specific universe. 

<br>

The extracted 3D radiomic features encompass several complementary categories:

- **First-order statistics:** global intensity-based descriptors (mean, variance, skewness, entropy).  
- **Shape features:** volumetric morphology metrics (sphericity, surface area, compactness, and volume).  
- **Texture features:**  
  - **GLCM:** Gray Level Co-occurrence Matrix  
  - **GLRLM:** Gray Level Run Length Matrix  
  - **GLSZM:** Gray Level Size Zone Matrix  
  - **NGTDM:** Neighboring Gray Tone Difference Matrix  
  - **GLDM:** Gray Level Dependence Matrix  
- **Filtered features:** derived from Laplacian of Gaussian (LoG) and wavelet-transformed images to capture multi-scale spatial patterns and textural heterogeneity.

#### CT Pre-processing
<a id="preprocessing"></a>
It is well known that CT images often contain significant noise and artifacts, which can hinder effective feature extraction, neither all CT regions are equally relevant to the diagnosis. To address these challenges, pre-processing techniques are essential for improving image quality. In this study, a sequential pre-processing pipeline was developed to enhance the overall quality of CT scans, consisting of the following steps:

- Body Segmentation
- Homogenous pixel spacing  
- HU widowing  
- Normalization  
- Filtering  

In order to choose the best filtering technique, the Peak Signal-to-Noise Ratio (PSNR) and the Structural Similarity Index (SSIM) were evaluated for each method. These quantitative metrics allowed for an objective comparison of image quality after filtering, ensuring that the selected technique provided optimal noise reduction while preserving important anatomical structures.

##### Body Segmentation

In an initial approach, body and lung segmentation were performed on CT slices from the LIDC dataset using intensity-based thresholding in the Hounsfield Unit (HU) domain. Body segmentation was achieved by isolating pixels with attenuation values above −500 HU, corresponding approximately to soft tissues and bones, while excluding air and background regions. The resulting binary mask was refined through morphological operations, including small object removal, binary closing, and hole filling, to eliminate noise and produce a continuous body region. The cleaned body mask was then applied to the CT image to exclude non-anatomical areas and improve the definition of the region of interest.

Subsequently, lung segmentation was attempted within the body mask by selecting pixels with intensities between −1000 HU and −400 HU, corresponding to pulmonary parenchyma and air-filled regions. Morphological filters were again applied to reduce artifacts and ensure the connectivity of the lung structures, and the two largest connected components were retained to represent the left and right lungs. However, despite these refinements, the lung segmentation process exhibited instability, as reflected in inconsistent region detection across slices and occasional omission of lung structures, as illustrated in the example below. This instability motivated the use of body segmentation alone for subsequent feature extraction to ensure reproducibility and robustness.

##### Homogeneous Pixel Spacing

This preprocessing technique ensures that each pixel represents the same physical distance in all directions. In medical imaging, scans from different patients or devices may have varying pixel spacing due to differences in acquisition protocols or equipment. If these differences are not corrected, measurements and features extracted from the images can be inconsistent or misleading. By resampling images to a uniform pixel spacing, we enable accurate quantitative analysis, fair comparison between scans, and reliable application of automated algorithms such as segmentation or radiomics. This step is essential for reproducibility and robustness in medical image processing workflows.

As an example, it was validated if CTs across different patient have different Pixel Spacing:

Since different patients have different Pixel Spacing, a function is used to rescale a 2D image so that its pixel spacing becomes isotropic (equal in both directions) by calculating the required zoom factors and applying interpolation with. The new isotropic space is 1x1 (mm).

##### HU Windowing

Hounsfield Unit (HU) windowing is a technique used to enhance the visualization of specific tissue types in CT images by selecting a relevant range of HU values. Each tissue in the body has a characteristic HU value (e.g., air is approximately -1000 HU). By applying an appropriate HU range, irrelevant structures and noise outside the desired interval are suppressed, improving the contrast and clarity of the anatomical regions of interest, such as lungs. This process facilitates more accurate segmentation and feature extraction in medical image analysis. For this study, a HU range of [-1200, 600] was applied to exclude undesired tissue densities effectively.

##### Normalization

The normalization function rescales the intensity values of each CT image slice to the [0, 1] range by subtracting the minimum value and dividing by the intensity range (maximum minus minimum). This step ensures that all images have a consistent intensity scale, which is essential for robust quantitative analysis and for the application of machine learning algorithms. By normalizing the data, variations due to acquisition parameters or scanner differences are minimized, facilitating fair comparison between images and improving the stability and convergence of subsequent processing steps


##### Filtering

Filtering is a crucial step in CT preprocessing as it helps to reduce noise and enhance the quality of the images, ensuring that important anatomical structures are preserved. CT scans often contain artifacts and random variations in pixel intensity that can obscure critical details, making it challenging to extract meaningful features. By applying appropriate filtering techniques, such as Gaussian, Median, Gabor, Adaptative Non-Local Means (ANLM), Block-Matching and 3D Filterin (BM3D) and Laplacian of Gaussian (LoG) filters, the signal-to-noise ratio is improved, facilitating more accurate feature extraction and analysis. This step is particularly important in medical imaging, where the clarity and precision of the data directly impact diagnostic accuracy and the reliability of automated algorithms.


| Filter          | Pros                                      | Cons                                      |  
|------------------|------------------------------------------|-------------------------------------------|  
| Gaussian  | Smooths noise effectively, easy to apply | May blur edges and fine details           |  
| Median    | Preserves edges while reducing noise     | Computationally expensive for large images|  
| Gabor     | Good for edge/texture, preserves structures | Computationally intensive, needs tuning   |  
| ANLM | Excellent noise reduction, preserves fine details | Very slow, high memory, needs noise estimate |  
| BM3D            | State-of-the-art denoising, preserves details | Computationally intensive, complex to implement |  
| LoG | Enhances edges, detects blobs effectively | Sensitive to noise, may over-enhance edges       |  

The PSNR and SSIM equations are as it follows:

**PSNR:**  
$$
PSNR = 10 \cdot \log_{10} \left( \frac{MAX_I^2}{MSE} \right)
$$  
Where $MAX_I$ is the maximum possible pixel value of the image and $MSE$ is the Mean Squared Error.  

**SSIM:**  
$$
SSIM(x, y) = \frac{(2\mu_x\mu_y + C_1)(2\sigma_{xy} + C_2)}{(\mu_x^2 + \mu_y^2 + C_1)(\sigma_x^2 + \sigma_y^2 + C_2)}
$$  
Where:

### Deep Learning-based Feature Extraction using DenseNet

In addition to the hand-crafted and radiomic features and preprocessing of the DICOM files, our pipeline incorporates a deep feature extraction stage using a pretrained DenseNet model. To do this, we implemented a custom **PyTorch class** which wraps the DenseNet backbone and exposes a .forward() interface accepting a single input image (slice), then outputs a feature vector.  

##### DenseNet vs. ResNet: A Comparative Analysis

The choice of DenseNet over other architectures, such as ResNet (a general-purpose baseline), is predicated on its superior efficiency, feature utilization, and inherent suitability for specialized datasets like those encountered in medical imaging. The primary justification is as follows:


##### Suitability for Specialized Datasets

DenseNet has demonstrated considerable efficacy in medical imaging contexts, even with relatively smaller datasets (e.g., chest X-rays). While LIDC-IDRI is a large public resource, it remains a specialized domain compared to massive general image datasets like ImageNet. DenseNet's capacity to capture subtle, fine-grained patterns makes it an optimal choice for the detailed classification tasks required for pulmonary nodules, where minor visual cues can be diagnostically significant.

In summary, while ResNet is a highly capable and robust general baseline, DenseNet's architectural strengths, including its parameter efficiency and optimal feature reuse, render it the preferred choice for this lung nodule classification task, directly supporting the project's goal of achieving high accuracy and robust generalization on specialized medical data.

##### Wich layers did we choose?

The specific combination of denseblock3, denseblock4, and norm5 offers a composite representation that effectively balances low-level detail with high-level abstraction:

- **denseblock3** : Provides a mid-to-high level representation that retains crucial spatial structure. It is instrumental in capturing local morphological information, which is vital for identifying texture and shape variations in pulmonary nodules.
- **denseblock4** : Represents the deepest convolutional block before final normalization. It encodes complex, abstract features that correlate with larger anatomical context and higher-order visual semantics.
- **norm5** : As the final batch normalization layer, it refines the aggregated representation from the entire feature hierarchy. It provides the most semantically complete feature map before task-specific classification, capturing global, high-level descriptors without the bias of the ImageNet classifier.

Together, these layers effectively leverage DenseNet’s feature reuse principle, yielding an optimal transfer of learned representations to the medical imaging domain.


##### Grayscale images vs RGB pretrained model  
Because our CT slices are naturally **single-channel (grayscale)**, whilst the DenseNet model was originally trained on three-channel RGB images, we adopt a pragmatic strategy: we replicate the grayscale slice across three channels (R=G=B) so that it conforms to the input shape the model expects. We justify this decision based on the following considerations:

- In many medical imaging settings, the difference between using RGB vs grayscale is relatively modest. For example, a recent study found that a U-Net model trained on grayscale CT images achieved segmentation performance comparable than a similarly configured RGB model ''[Unintended Bias in 2D+ Image Segmentation and Its Effect on Attention Asymmetry](https://arxiv.org/html/2505.14105v2)''.
- The critical features in our application are derived from structural and intensity variations of the lesion, not color/hue information hence duplicating the grayscale channel into three inputs does not meaningfully lose signal.  

Thus, although the network was optimized for RGB natural-images, the adaptation to grayscale medical slices is well justified and supported by precedent in the literature ''[ Enhancing Chest X-Ray Image Classiﬁcation for Lung Diseases through Machine Learning ](https://www.researchgate.net/publication/378160884_Enhancing_Chest_X-Ray_Image_Classification_for_Lung_Diseases_through_Machine_Learning)''. 


##### Parallel execution per patient  

The processing flow is as follows:  
1. We iterate **in parallel over patients** (i.e., each patient’s scan is handled concurrently via a multiprocessing or thread pool mechanism).  
2. For each patient, we identify all slices that contain the nodule (based on the segmentation mask from PyLIDC).  
3. Each such slice is passed through our DenseNet wrapper class to extract a feature vector.  
4. We store, for each extracted slice, a record in a CSV file containing the keys: `patient_id`, `nodule_id`, `slice_id`, and the feature vector (flattened).

- **Nodule-level analysis** - simplifies the data for classical machine learning benchmarks, where each nodule is represented by a single, aggregated feature vector.

### Bag of Features (BoF) for Pulmonary Nodules

#### Overview

The Bag of Features (BoF), also referred to as Bag of Visual Words (BoVW), is a representation technique adapted from the Bag of Words model in natural language processing.  
In the context of **pulmonary nodule classification**, BoF converts variable-size sets of nodule features, such as **slice-level DenseNet activations** into a **fixed-length statistical representation** that characterizes the distribution of local patterns across the nodule volume.  
This approach is particularly useful when the **spatial arrangement of features within a nodule is secondary** to the overall texture or appearance statistics.

#### How it works

The BoF pipeline for pulmonary nodules typically consists of the following steps:

1. **Feature detection / sampling**  
   - Each nodule is represented by a set of **slice-level deep features** descriptors extracted previously.  
   - Local “patches” or slices serve as the units of analysis, capturing texture, intensity, and shape variations within the nodule.

2. **Feature description**  
   - Each slice or patch is summarized by its **feature vector**, e.g., DenseNet activations.

3. **Codebook (visual vocabulary) construction**  
   - Feature vectors from a **representative set of nodules** are clustered via **k-means** (or another clustering algorithm) to form a finite set of prototype vectors called **visual words**.  
   - The set of cluster centers constitutes the **nodule codebook**, encoding common patterns across the dataset.

4. **Quantization (feature assignment)**  
   - Each slice-level descriptor of a nodule is mapped to the nearest codebook vector (hard assignment) or multiple nearby vectors (soft assignment), converting a variable-size set of descriptors into **discrete occurrences of visual words**.

5. **Histogram aggregation**  
   - Visual word occurrences are aggregated across all slices of the nodule to create a **fixed-length histogram**, representing the nodule as a whole.  
   - This is equivalent to averaging the slice-level BoF representations into a **nodule-level descriptor**, as applied in our dataset.

6. **Classification / downstream analysis**  
   - The resulting fixed-length nodule descriptor can be used as input to conventional machine learning classifiers (e.g., SVM, Random Forest, XGBoost) to predict **benign vs malignant** nodules.  
   - Alternatively, these histograms can support similarity-based retrieval or exploratory data analysis.

#### Strengths and limitations

**Strengths:**

- Produces fixed-length descriptors from variable-size nodule slices.  
- Enables fusion of heterogeneous features into a single nodule-level representation.  
- Interpretable and suitable for small-to-moderate datasets.

**Limitations:**

- Discards precise spatial arrangement of features within the nodule unless augmented with spatial pyramids.  
- Sensitive to codebook size, descriptor selection, and quantization method.  
- On large datasets, end-to-end deep learning may outperform BoF, but BoF remains valuable for interpretability and feature benchmarking.

This section describes the EDA methodology used to prepare feature sets for downstream analysis and dimensionality reduction. The primary goal was to identify and retain the most informative features while optimizing data quality.

**DenseNet Features Pipeline:**
1. **Low-variance filter** - Removed features with minimal variance that contribute little discriminative power
2. **Correlation clustering** - Grouped highly correlated features to address multicollinearity and redundancy
3. **Standard normalization** - Applied StandardScaler to ensure equal contribution of all features
4. **PCA** - Transformed features into uncorrelated principal components while retaining 99.9% of variance

**Radiomics Features Pipeline:**
1. **Correlation clustering** - Managed feature redundancy and selected representative subsets
2. **Robust scaler** - Applied RobustScaler using median and IQR to handle outliers effectively

This differentiated approach ensures each feature type is optimally prepared for advanced analytical techniques, with DenseNet features undergoing more aggressive dimensionality reduction while radiomics features maintain clinical interpretability.


##### Low Variance Feature Selection

The main objective of this selection is to identify and remove features with very low variance (little information) to reduce noise and computational cost, using data-relative criteria.

Methodology:
- Compute the mean per feature:
    $$\bar{x}_i = \frac{1}{N}\sum_{j=1}^N x_{ji}$$
- Compute the variance per feature:
    $$
    \mathrm{var}_i = \frac{1}{N}\sum_{j=1}^N (x_{ji} - \bar{x}_i)^2
    $$
- Visualize the distribution of variances (histogram / cumulative curve) to choose a conservative cutoff.
- Define a relative cutoff, for example:
    $$
    \text{cutoff} = \alpha \times \mathrm{median}_i(\mathrm{var}_i)
    $$
    $$\text{with } \alpha = 0.01 \text{ --> remove features with } \mathrm{var}_i < \text{cutoff}$$

**Note on Radiomics:** do not apply a low-variance filter directly to radiomic features, since they have highly heterogeneous scales and domains, a small absolute value can be informative in its own domain. Instead, use selection methods that respect scale and interpretability (e.g., correlation clustering, importance-based selection, or robust scaling prior to filtering).


### Correlation Clustering for Feature Redundancy Reduction

##### The Adapted Complete-Linkage Correlation Method

To address feature redundancy, an adapted complete-linkage clustering method based on a strict correlation threshold was implemented. This method operates on a fundamental principle:

A feature is incorporated into an existing cluster exclusively if its absolute correlation with every single feature already present in that cluster meets or exceeds a predefined threshold (e.g., 0.9).

This stringent criterion ensures the formation of highly cohesive and internally consistent clusters. Within each cluster, every feature is guaranteed to exhibit a strong correlation with all other features. If a feature fails to satisfy this condition for any existing cluster, it initiates the formation of a new, distinct cluster.

##### Evaluation of Ward's Method

Initially, **Ward's method** was considered for feature redundancy reduction because its objective of minimizing the increase in total within-cluster variance upon merging offered a promising approach for creating compact and balanced feature groups. This method works by identifying and combining the pair of clusters that, when merged, result in the smallest increase in the overall variation within the clusters. Unlike other linkage methods (e.g., single or complete linkage), Ward's method focuses on the compactness and similarity of data points within groups, thereby tending to produce clusters that are more balanced and uniform in size.

However, two significant limitations precluded the application of Ward's method in this context:

- **Variance-Based Merge Criterion**: As previously discussed in the context of low variance filtering, a variance-based merge criterion is unsuitable for radiomic features. The disparate dimensionalities and scales of radiomic features render variance an unreliable metric for assessing feature importance or redundancy across the entire feature set.

- **Chaining Effect**: Ward's method is susceptible to a "chaining effect." This phenomenon can occur when a feature C joins a cluster {A, B} even if the correlation between A and C is low, provided that the overall increase in variance remains minimal. This can lead to the formation of "chains" where features A and D are grouped into the same cluster due to their linkage through intermediary features B and C, despite a potentially low correlation between A and D. Such an effect can result in a loss of informational clarity and the grouping of genuinely dissimilar features.

By employing the adapted complete-linkage correlation method, these issues are effectively avoided. The interpretability of the feature clusters is significantly enhanced, as the clustering logic is transparently based on a direct and easily understood correlation threshold.


### Feature Normalization and Scaling

#### Rationale for Normalization

Prior to the integration of DenseNet activations and PyRadiomics features, normalization is a critical preprocessing step.

Without proper normalization, features with significantly larger numerical ranges can disproportionately influence distance calculations, correlation analyses, and PCA components, thereby introducing an undesirable bias into the overall analysis. However, it is crucial to recognize that Convolutional Neural Network (CNN) activations and radiomic features originate from fundamentally different statistical distributions, which necessitates distinct scaling approaches rather than a uniform scaler.

#### StandardScaler for DenseNet Features

The StandardScaler was selected for DenseNet features due to its ability to transform each feature to possess a mean of 0 and a standard deviation of 1. 
In essence, StandardScaler preserves the statistical integrity of DenseNet embeddings while harmonizing all features onto a common scale, an indispensable prerequisite before applying dimensionality reduction or merging with other normalized datasets.

#### RobustScaler for Radiomic Features

The RobustScaler was chosen for radiomic features as it standardizes data using the median and the interquartile range (IQR), rather than the mean and standard deviation.

Its selection is justified by the following considerations:


- It effectively mitigates the impact of outliers without introducing distortion to the majority of the data.
- It balances features with disparate numerical scales, preventing any single feature from unduly influencing the analysis merely due to its magnitude.
- It preserves the relative relationships between features and avoids the generation of missing or invalid values.

##### Rationale Against Log Transformation

Initially, a logarithmic transformation (log1p) was considered to mitigate data skewness. However, this transformation is only valid for strictly positive numbers. Given that the dataset includes zeros and negative values, applying log1p would have resulted in the generation of ``NaN`` values and a consequent loss of information. To maintain the validity and integrity of all features, the log transform was ultimately deemed unsuitable.

### Dimensionality Reduction with Principal Component Analysis (PCA)

Following feature harmonization, Principal Component Analysis (PCA) was exclusively applied to the DenseNet features.

**Rationale for PCA Post-Normalization**

PCA operates under the assumption that all features are on comparable scales; otherwise, components with higher variance would disproportionately dominate the decomposition. The preceding normalization step (using StandardScaler for DenseNet features) ensures that each neuron (feature) contributes proportionally to the PCA process.

PCA identifies orthogonal directions, termed "principal components," that maximize the variance within the data. Mathematically, this is represented as Z = XW, where Z denotes the matrix of principal components, X is the normalized feature matrix, and W comprises the eigenvectors of the covariance matrix of X. Each resulting PCA feature (principal component) is a linear combination of the original features, effectively serving as a new, compressed representation that encapsulates the majority of the original information.

![pca](EDA1\pca_eda_plots\pca_explained_variance.png)

The decision to select principal components that explain 99.9% of the total variance, rather than a lower threshold (e.g., 95%), represents a conservative approach. This ensures that no potentially informative patterns, particularly subtle yet relevant variations often distributed across numerous neurons within DenseNet embeddings, are inadvertently discarded at this preliminary stage of dimensionality reduction.

**Rationale Against Applying PCA to PyRadiomics Features**

We deliberately refrained from applying PCA to the radiomics dataset for several compelling reasons:

- Explainability: Each radiomics feature possesses a defined clinical or physical interpretation (e.g., related to texture, shape, or intensity). Applying PCA would transform these interpretable features into abstract, orthogonal components, thereby rendering the interpretation of individual new dimensions exceptionally challenging or impossible in a clinically meaningful context.
- Transparency: Maintaining the original radiomics features allows for direct comparison and straightforward interpretation of which specific radiomic characteristics contribute most significantly to downstream analytical tasks and predictive models.

In summary, PCA was applied solely to the DenseNet feature set to effectively reduce redundancy while preserving nearly all the original variance. Conversely, radiomics features were retained in their original, interpretable form to uphold transparency and clinical relevance.

**Why Variance Filter and Correlation Clustering Preceded PCA**

Applying variance filtering and correlation clustering before PCA optimized dimensionality reduction by removing low-variance and highly correlated features, reducing noise and redundancy. With this, we ensure that these processes are carried out effectively, improving PCA’s efficiency, robustness, and interpretability, and leading to more meaningful and independent principal components for subsequent modeling.

```python
import pandas as pd
import io
from collections import Counter

def _rank_single_malignancy_type(df_input):
    
    df_filtered = df_input.copy()
    df_filtered.loc[:, "dataset_count"] = df_filtered["dataset"].apply(
        lambda x: len(x.split("||")) if isinstance(x, str) and "||" in x else 1
    )
    ranked_dataframes_for_type = {}


    for count in range(1, 8):
        count_df = df_filtered[df_filtered["dataset_count"] == count].copy()
        
        if not count_df.empty:
            count_df_sorted = count_df.sort_values(
                by=["test_roc_auc_ovr", "test_accuracy", "test_recall"], 
                ascending=[False, False, False]
            )
            
            top_3_results = count_df_sorted.head(3).copy()
            top_3_results.insert(0, 'Rank', range(1, len(top_3_results) + 1))
            output_df = top_3_results[['Rank', 'dataset', 'test_roc_auc_ovr', 'test_accuracy', 'test_recall', 'modelo']].copy()
            output_df.columns = ['Rank', 'Dataset', 'ROC-AUC', 'Accuracy', 'Recall', 'Model']
            output_df['ROC-AUC'] = output_df['ROC-AUC'].round(2)
            output_df['Accuracy'] = output_df['Accuracy'].round(2)
            output_df['Recall'] = output_df['Recall'].round(2)
            output_df = output_df.reset_index(drop=True)

            ranked_dataframes_for_type[count] = output_df
            
        else:
            ranked_dataframes_for_type[count] = pd.DataFrame(columns=['Rank', 'Dataset', 'ROC-AUC', 'Accuracy', 'Recall', 'Model'])
        
    return ranked_dataframes_for_type

def get_ranked_datasets_as_dfs_by_malignancy(csv_file_path):
    df = pd.read_csv(csv_file_path)
    df = df[df["modelo"].apply(lambda x: isinstance(x, str) and not str(x).replace(".", "").isdigit())]
    df = df[df["bin_malignancy"].isin(['True', 'False'])].copy()
    df["bin_malignancy"] = df["bin_malignancy"] == 'True'

    all_results_by_malignancy = {}

    df_binary = df[df["bin_malignancy"] == True].copy()
    df_multiclass = df[df["bin_malignancy"] == False].copy()

    all_results_by_malignancy["binário"] = _rank_single_malignancy_type(df_binary)
    all_results_by_malignancy["multi-classe"] = _rank_single_malignancy_type(df_multiclass)
        
    return all_results_by_malignancy

if __name__ == "__main__":
    
    csv_path = r"modelos\resultados_modelos.csv"
    
    df_check = pd.read_csv(csv_path)
    df_check = df_check[df_check["bin_malignancy"].isin(['True', 'False'])].copy()
    df_check["bin_malignancy"] = df_check["bin_malignancy"] == 'True'
    df_check['dataset_count'] = df_check["dataset"].apply(
        lambda x: len(x.split("||")) if isinstance(x, str) and "||" in x else 1
    )
    pd.set_option('display.max_colwidth', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_columns', None)
    
    all_ranked_dataframes = get_ranked_datasets_as_dfs_by_malignancy(csv_path)
    
    for count in range(1, 8):
        count_df = df_check[df_check['dataset_count'] == count]
        unique_datasets_binary = count_df[count_df['bin_malignancy'] == True]['dataset'].nunique()
        unique_datasets_multi = count_df[count_df['bin_malignancy'] == False]['dataset'].nunique()

    
    for malignancy_type, ranked_dfs_for_type in all_ranked_dataframes.items():
        print(f"## Rankings para {malignancy_type}:\n")
        for count, df in ranked_dfs_for_type.items():
            num_results = len(df)
            print(f"### Ranking para grupos com {count} dataset(s):\n")
            if not df.empty:
                display(df)
            else:
                print("Nenhum resultado encontrado para este grupo.\n")
            print("\n")
```

Among all, the **SVM consistently dominated the benchmarks**, achieving the best performance across nearly all datasets and combinations.

#### Exclusion of Hyperparameter Tuning and Cross-Validation

Due to the **high computational cost** associated with testing hundreds of feature combinations across multiple classifiers, both **hyperparameter tuning** and **cross-validation** were deliberately excluded from this benchmarking phase.  
Instead, **default configurations** were used to provide a fair and consistent baseline comparison of model behavior.  
This approach ensures that performance differences reflect **intrinsic feature quality and dataset composition**, rather than extensive model optimization.

#### validation and test scores
The performance of the classification models was evaluated using both **validation** and **test** datasets.  
Interestingly, the results on validation and test sets are **very similar**, indicating several important points:

- **Good generalization:** The similarity suggests that the models are not overfitting to the training data and that the extracted features (radiomic and deep) provide consistent, discriminative information across unseen samples.  
- **Dataset representativeness:** It implies that the distribution of nodules in the validation and test sets is similar, which supports the robustness of our sampling and preprocessing strategy.  
- **Feature stability:** Features that perform well on validation tend to retain their predictive power on test, highlighting the relevance of the selected descriptors.


#### Frequency Analysis of Individual and Combined Feature Sets

A detailed frequency analysis was performed to understand which individual feature sets are most frequently included in the **top-performing combined datasets**. Key observations include:

### Final Discussion and Future Work

Despite the creative effort to apply the **Bag of Features (BoF)** methodology to pulmonary nodule analysis, the results did not meet the initial expectations.  
While the concept aimed to capture intra-nodule variability by aggregating **slice-level representations** (derived from DenseNet features) into **nodule-level descriptors**, the final performance lagged behind the classical feature fusion approaches tested earlier.

Several factors may have contributed to this outcome:

- **Information loss during quantization:**  
  The transformation of continuous feature vectors into discrete “visual words” through clustering may have discarded subtle but relevant differences between nodules, which are critical in medical imaging contexts.

- **Insufficient spatial context:**  
  The BoF model intentionally ignores the spatial arrangement of features, which in natural images is often tolerable, but in medical CT data - where lesion structure and internal gradients carry diagnostic meaning this simplification likely limited performance.


Nevertheless, the experiment demonstrates **an innovative and exploratory approach** that bridges classical computer vision.  
The idea of representing a 3D nodule as a “distribution of local slice-level visual words” remains conceptually appealing and could be valuable if refined with modern adaptations such as:

- **Spatially-aware BoF extensions** (e.g., spatial pyramids or 3D voxel clustering) to retain geometric information.  
- **Deep embedding clustering** to generate more meaningful, semantically aligned codebooks.  
- **Hybrid pipelines** combining BoF histograms with deep feature aggregation strategies like NetVLAD or attention-based pooling.

Although the BoF implementation did not yield competitive accuracy in this study, it illustrates the value of **creative experimentation** and opens a direction for **future research** potentially improving interpretability and feature compactness in lung nodule classification tasks.
